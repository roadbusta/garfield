Read in locations as DataFrame

```python
import pandas as pd

df = pd.read_csv("../data/cleaned/locations.csv")

df.head()
```

Load in points into plotly dash map

```python
df["step"] = range(len(df))
```

```python
import plotly.express as px

fig = px.scatter_map(
    df,
    lat="lat",
    lon="lon",
    animation_frame="step",   #  slider control
    zoom=15,
)

# Add a connected line that grows over time
fig.update_traces(mode="markers+lines")

fig.update_layout(
    map_style="carto-positron",
    margin=dict(r=0, t=0, l=0, b=0)
)

fig.show()
```

```python
import folium
import pandas as pd



# Create the base map centered at the first point (adjust zoom level as needed)
m = folium.Map(location=[df["lat"].iloc[0], df["lon"].iloc[0]], zoom_start=15)

# Loop through the dataframe and add custom emoji markers
for i, row in df.iterrows():
    folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=f"<b>{row['time']}</b>",  # Add a popup with timestamp (or any other info)
        icon=folium.DivIcon(html=f'<div style="font-size: 24px; color: black;"></div>')  # Cat emoji
    ).add_to(m)

# Save the map as an HTML file
# m.save("emoji_map.html")

# If running in a Jupyter notebook, you can display the map directly:
m
```
